```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
n, N = 40, 60
A = np.random.randn(n, N)
plt.matshow(A);
```

```python
from scipy import linalg
```

```python
U, s, Vh = linalg.svd(A)
```

```python
U.shape, s.shape, Vh.shape
```

```python
S = np.zeros((n, N))
for k in range(40):
    S[k, k] = s[k]

plt.matshow(S);
```

SVD:

$A = U S V^\top$

avec $U^\top U = I$ et $V^\top V = I$ et S diagonale

```python
U.shape, S.shape, Vh.shape
```

```python
A - np.dot(U, np.dot(S, Vh))
```

```python
plt.matshow(np.dot(Vh, Vh.T))
plt.colorbar()
```

```python
x0 = np.zeros(N)
x0[10] = 30
x0[44] = -20
b = np.dot(A, x0)
```

```python
plt.stem(x0);
```

```python
b.shape
```

```python
lambd = 100.

def f(x):
    return 0.5 * np.linalg.norm(b - np.dot(A, x)) ** 2

def pobj(z):
    return f(x) + lambd * np.sum(np.abs(x))

def grad_f(z):
    return - np.dot(A.T, b - np.dot(A, x))
            
def prox(x, rho):
    """Minimize
    
    0.5  * || x - y ||^2 + rho * lambd || x ||_1
    
    """
    return np.sign(x) * np.maximum(np.abs(x) - rho * lambd, 0)

x = np.zeros(N)
rho = 0.001

all_pobj = []

for k in range(300):
    x = x - rho * grad_f(x)
    x = prox(x, rho)
    all_pobj.append(pobj(x))

f , (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
ax1.plot(all_pobj);
ax2.stem(x0, linefmt='r-', markerfmt='ro', basefmt='r-');
ax2.stem(x, linefmt='b-', markerfmt='bo', basefmt='r-');
```

```python
xx = np.linspace(-10, 10, 100)

plt.plot(xx, prox(xx, 0.05));
plt.plot(xx, xx, 'r');
```
