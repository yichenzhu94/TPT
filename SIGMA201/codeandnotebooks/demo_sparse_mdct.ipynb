```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
from scipy.io import wavfile
```

```python
from stft import stft, istft

Fe, s = wavfile.read('glockenspiel.wav')
print(s.shape)
```

```python
plt.plot(s)
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude');
```

```python
plt.hist(s, bins=30);
```

```python
import IPython
IPython.display.Audio(s, rate=Fe)
```

### Compute the STFT

```python
wsize = 32 * 8
s_stft = stft(s, wsize=wsize)[0]
```

```python
print(s_stft.dtype)
print(s_stft.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

ax1.imshow(20 * np.log10(np.abs(s_stft)), origin='lower', cmap=plt.cm.viridis)
ax1.set_xlabel('Frame')
ax1.set_ylabel('Frequency')
ax1.axis('tight')

ax2.hist(np.abs(s_stft.ravel()), bins=20);
```

Test reconstruction

```python
s_inv_stft = istft(s_stft)
```

```python
print(s_inv_stft.shape)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
ax1.plot(s_inv_stft)
ax1.set_xlabel('Time (samples)')
ax1.set_ylabel('Amplitude');
ax2.plot(s_inv_stft - s)
ax2.set_xlabel('Time (samples)')
ax2.set_ylabel('Amplitude');
```

```python
IPython.display.Audio(s_inv_stft, rate=Fe)
```

### Now with an MDCT

```python
from mdct import mdct, imdct
```

```python
wsize = 32 * 4
s_mdct = mdct(s, wsize)
```

```python
print(s_mdct.shape)
```

```python
print(s_mdct.shape[0] * s_mdct.shape[1], s.size)
```

```python
print(s_mdct.dtype)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

ax1.imshow(20 * np.log10(np.abs(s_mdct)), origin='lower', cmap=plt.cm.viridis)
ax1.set_xlabel('Frame')
ax1.set_ylabel('Frequency')
ax1.axis('tight')

ax2.hist(s_mdct.ravel(), bins=20);
```

```python
s_inv_mdct = imdct(s_mdct, wsize)
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
ax1.plot(s_inv_mdct);
ax1.set_xlabel('Time (samples)')
ax1.set_ylabel('Amplitude');
ax2.plot(s_inv_mdct - s);
ax2.set_xlabel('Time (samples)')
ax2.set_ylabel('Amplitude');
```

```python
IPython.display.Audio(s_inv_mdct, rate=Fe)
```

### Visualizing the atoms of the dictionaries

```python
dummy_mdct = np.zeros_like(s_mdct)
dummy_mdct[10, 5] = 1.
atom = imdct(dummy_mdct, wsize)
plt.plot(atom[:1000]);
```
